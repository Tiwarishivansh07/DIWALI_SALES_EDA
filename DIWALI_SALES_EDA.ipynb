```python
import numpy as numpy
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df= pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.tail()
```

```python
df.info()
```

```python
pd.isnull(df)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns= {'Marital_Status': 'Vivah'})
```

```python
df.describe()
```

```python
df.columns
```

```python
a=sns.countplot(x = 'Gender', data=df)
for bars in a.containers: # this is used to store values in container and show the count in chart
    a.bar_label(bars)
```

```python
df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sale_bar=df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sale_bar)
```

```python
df.columns
```

```python
a=sns.countplot(x = 'Age Group',hue='Gender', data=df)
for bars in a.containers: # this is used to store values in container and show the count in chart
    a.bar_label(bars)
```

```python
sale_bar=df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sale_bar)
```

```python
sale_state=df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Orders',data=sale_state)
```

```python
sale_state=df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Amount',data=sale_state)
```

```python
a=sns.countplot(x = 'Marital_Status', data=df)
for bars in a.containers: # this is used to store values in container and show the count in chart
    a.bar_label(bars)
```

```python
marital_st=df.groupby(['Marital_Status','Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Marital_Status',y='Amount', hue='Gender', data=marital_st)
```

```python
sns.set(rc={'figure.figsize':(25,5)})
a=sns.countplot(x = 'Occupation', data=df)
for bars in a.containers: # this is used to store values in container and show the count in chart
    a.bar_label(bars)
```

```python
high_occu=df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Occupation',y='Amount', data=high_occu)
```

```python
a=sns.countplot(x = 'Product_Category', data=df)
for bars in a.containers: # this is used to store values in container and show the count in chart
    a.bar_label(bars)
```

```python
prod_cat=df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_Category',y='Amount', data=prod_cat)
```

```python
top_sel=df.groupby(['Product_ID'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(x='Product_ID',y='Amount', data=top_sel)
```
